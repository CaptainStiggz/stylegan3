```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
%matplotlib inline

file = "5a/ST05_SE010122.jpg"
#file = "5a/ST05_SE030907.jpg"
# file = "5a/ST05_SE030916.jpg" # pale yellow
# file = "51/ST05_SE010107.jpg"
image = cv2.imread(file)
print('This image is:', type(image), ' with dimensions:', image.shape)

image_copy = np.copy(image)
plt.imshow(image_copy)
```

```python
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)
plt.imshow(image_copy)
```

```python
# imghsv = cv2.cvtColor(image_copy, cv2.COLOR_RGB2HSV).astype("float32")
# (h, s, v) = cv2.split(imghsv)
# s = s*2
# # s = np.clip(s,0,255)
# imgmer = cv2.merge([h,s,v])
# imgsat = cv2.cvtColor(imgmer.astype("uint8"), cv2.COLOR_HSV2RGB)
# plt.imshow(imgsat)
imgsat = image_copy
```

```python
lower_blue = np.array([205,205,205])  
upper_blue = np.array([255,255,255])

mask = cv2.inRange(imgsat, lower_blue, upper_blue)
plt.imshow(mask, cmap='gray')
```

```python
def bbox_crop(img):
#     inverted = cv2.bitwise_not(img)
#     plt.imshow(inverted, cmap='gray')
#     rows = np.any(inverted, axis=1)
#     cols = np.any(inverted, axis=0)
#     rmin, rmax = np.where(rows)[0][[0, -1]]
#     cmin, cmax = np.where(cols)[0][[0, -1]]
#     print(rmin, rmax)
#     print(cmin, cmax)
#     print(img[rmin:rmax, cmin:cmax].shape)
#     return img[rmin:rmax, cmin:cmax]
#     img = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
    contours, hierarchy = cv2.findContours(img, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)[-2:]
    idx = 0 
    pad = 50
    for cnt in contours:
        idx += 1
        x,y,w,h = cv2.boundingRect(cnt)
        if w > 100 and h > 100:
            xx,yy,ww,hh = x, y, w, h
            diff = w - h
            print(diff)
            if diff > 0:
                xx -= pad
                ww += 2 * pad
                yy = y - diff // 2 - pad
                hh = ww
            elif diff < 0:
                yy -= pad
                hh += 2 * pad
                xx = x + diff // 2 - pad
                ww = hh
            roi=img[yy:yy+hh,xx:xx+ww]
            return roi,xx,yy,ww,hh

cropped, x, y, w, h = bbox_crop(mask)
print(cropped.shape)
plt.imshow(cropped, cmap='gray')
```

```python
masked_image = np.copy(image_copy)
masked_image[mask != 0] = [255, 255, 255]
plt.imshow(masked_image)
```

```python
print(cropped.shape)
final_cropped = masked_image[y:y+h,x:x+w]
plt.imshow(final_cropped)
```

```python
cv2.imwrite("output.jpg", cv2.cvtColor(masked_image, cv2.COLOR_RGB2BGR))
```
