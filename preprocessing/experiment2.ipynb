```python
import matplotlib.pyplot as plt
from process_v2 import Processor
import cv2
import numpy as np
%matplotlib inline
%load_ext autoreload
%autoreload 2
```

```python
processor = Processor()
# img = processor.rand_image("5a")
img = processor.read_file("5a/ST05_SE051899.jpg")
plt.imshow(img)
```

```python
gray = processor.threshold(img)
plt.imshow(gray, cmap='gray')
```

```python
contrast = processor.contrast(img, (3, 8))
plt.imshow(contrast)
```

```python
# find the largest contour area in the image
contour = processor.find_largest_contour(gray)
image_contour = np.copy(img)
cv2.drawContours(image_contour, [contour], 0, (0, 255, 0), 2, cv2.LINE_AA, maxLevel=1)
plt.imshow(image_contour)
```

```python
mask3d, mask2 = processor.create_contour_mask(gray, contour)
plt.imshow(mask3d)
```

```python
# create the foreground image by zeroing out the pixels where `mask2`...
# ... has black pixels
foreground = processor.apply_contour_mask(img, mask2)
plt.imshow(foreground)
```

```python
bbox = processor.bbox_contour(contour)
cropped = processor.crop(foreground, bbox)
plt.imshow(cropped)
```
